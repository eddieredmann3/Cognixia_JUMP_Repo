Part 3

```python
sns.boxplot(data=income_df, x="age", y="income")
```

```python
income_df["income"] = income_df["income"].replace(["<=50K", ">50K"], [0, 1])
```

```python
income_df.head()
```

```python
sns.set(rc={"figure.figsize":(21, 10.5)})
sns.countplot(data=income_df, x="age", hue="income")
```

```python
income_df.groupby(by="workclass").mean()["age"]
```

```python
income_df.groupby(by="gender").mean()["income"]
```

```python
income_df.head()
```

```python
race_df = income_df.groupby(by="race", as_index=False).mean()
race_df
```

```python
sns.set(rc={"figure.figsize":(21, 10.5)})
sns.barplot(data=race_df, x="race", y="income")
```

```python
marriage_df = income_df.groupby(by="marital-status", as_index=False).mean()
```

```python
sns.set(rc={"figure.figsize":(21, 10.5)})
sns.barplot(data=marriage_df, x="marital-status", y="income")
```

```python
hours_df = income_df.groupby(by="hours-per-week", as_index=False).mean()
```

```python
sns.set(rc={"figure.figsize":(21, 10.5)})
sns.barplot(data=hours_df, x="hours-per-week", y="income")
```
