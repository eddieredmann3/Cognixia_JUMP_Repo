```python
#imported libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset and first ten elements
comm_df = pd.read_csv("./Ecommerce_purchases.csv")
comm_df.head(10)
```

```python
#last 10 elements
comm_df.tail(10)
```

```python
#column datatypes
comm_df.dtypes
```

```python
#sorting values by purchase price
comm_df.sort_values(by="Purchase Price")
```

```python
#same sort, but descending
comm_df.sort_values(by="Purchase Price", ascending=False)
```

```python
#finding the mean purchase price
comm_df.agg({"Purchase Price": ['mean']})
```

```python
#looking at the number of orders by language, French has 1097 orders
lang_count = comm_df.groupby(by="Language", as_index=False)["Address"].count()
lang_count
```

```python
#looking for all the orders placed by engineers
comm_df[comm_df["Job"].str.contains("Engineer", case=False)]
```

```python
#looking for the email of a person with a specific email address
comm_df[comm_df["IP Address"] == "132.207.160.22"]["Email"]
```

```python
#all purchases over $50 made with a Mastercard
comm_df[(comm_df["CC Provider"] == "Mastercard") & (comm_df["Purchase Price"] >= 50)]
```

```python
#looking for a person's email with the credit card number specified
comm_df[comm_df["Credit Card"] == 4664825258997302]["Email"]
```

```python
#looking for the number of people who purchased in the morning
comm_df[comm_df["AM or PM"] == "AM"].shape
```

```python
# and in the evening
comm_df[comm_df["AM or PM"] == "PM"].shape
```

```python
#looking for everyone whose credit card expired in 2020
comm_df[comm_df["CC Exp Date"].str.contains("/20")]
```

```python
#separatting everyone's email into name and domain so that we can find the largest domains
comm_df[["Email_Name", "Email_Domain"]] = comm_df["Email"].str.split("@", expand=True)
```

```python
#finding the largest domains
comm_df.groupby(["Email_Domain"])["Email_Name"].count().nlargest()
```

Personal Insights

```python
#I wanted to find out which browsers were most popular, but the information isn't uniform
comm_df[["Browser_Info_Cleaned", "OS_Info"]] = comm_df["Browser Info"].str.split("/", n=1, expand=True)
```

```python
# you can see the lack of uniformity between Opera and Firefox
comm_df[["Browser_Info_Cleaned", "OS_Info"]]
```

```python
#evidently nobody used Chrome or Safari. Weird
comm_df.groupby(by="Browser_Info_Cleaned")["Email"].count()
```

```python
#looking to see if the purchase price info is a normal distribution
sns.histplot(data=comm_df, x="Purchase Price", bins=10)
```

```python
#looking for the biggest companies
comm_df.groupby(["Company"])["Email_Name"].count().nlargest()
```

```python
#seeing if there were any users who ordered a lot more than others
comm_df.groupby(["Email_Name"])["Email_Domain"].count().nlargest()
```

```python
#looking at the jobs that purchased the most
comm_df.groupby(["Job"])["Email_Domain"].count().nlargest()
```

```python
#Looking at the most popular type of credit card
comm_df.groupby(["CC Provider"])["Email_Domain"].count().nlargest(10)
```
