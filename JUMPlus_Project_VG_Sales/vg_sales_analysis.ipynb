```python
vg_sales_new_data = pd.read_csv("./refined_vg_sales_analysis.csv")
```

```python
vg_sales_new_data.head()
```

```python
release_year_data = vg_sales_new_data.groupby(by=["Year_of_Release"]).sum()
release_year_data
```

```python
sns.set(rc={"figure.figsize":(19, 10.5)})
sns.barplot(data=release_year_data, x=release_year_data.index, y="NA_Sales")
```

```python
sns.barplot(data=release_year_data, x=release_year_data.index, y="EU_Sales")
```

```python
sns.barplot(data=release_year_data, x=release_year_data.index, y="JP_Sales")
```

```python
sns.barplot(data=release_year_data, x=release_year_data.index, y="Other_Sales")
```

```python
genre_sales = vg_sales_new_data.groupby(by=["Genre"]).sum()
genre_sales
```

```python
sns.set(rc={"figure.figsize":(19, 10.5)})
sns.barplot(data=genre_sales, x=genre_sales.index, y="NA_Sales")
```

```python
sns.set(rc={"figure.figsize":(19, 10.5)})
sns.barplot(data=genre_sales, x=genre_sales.index, y="EU_Sales")
```

```python
sns.set(rc={"figure.figsize":(19, 10.5)})
sns.barplot(data=genre_sales, x=genre_sales.index, y="JP_Sales")
```

```python
sns.set(rc={"figure.figsize":(19, 10.5)})
sns.barplot(data=genre_sales, x=genre_sales.index, y="Other_Sales")
```

```python
platform_sales = vg_sales_new_data.groupby(by=["Platform"]).sum()
platform_sales
```
